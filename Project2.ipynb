```python
# Modelling
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay, classification_report
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
```

```python

x = df[['Age','Num_Lab_Procedures','Num_Medications','Num_Diagnoses','Num_Outpatient_Visits','Num_Inpatient_Visits','Num_Emergency_Visits']] 
```

```python
y = df['Readmitted']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=32)
```

```python
# Create a Random Forest Classifier
random_forest_model = RandomForestClassifier(n_estimators=100,max_depth=20,random_state=32)
```

```python
random_forest_model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = random_forest_model.predict(X_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
class_report = classification_report(y_test, y_pred, target_names=['Yes','No'])
print("Classification Report:\n", class_report)
```
